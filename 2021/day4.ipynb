```python
with open("src/day4/input.txt") as fileobject:
    lines = fileobject.read().splitlines()
    bingo_order = lines[0]
    cards = []
    blank_space_position = 1
    
    while blank_space_position < len(lines):
        card = lines[blank_space_position+1:blank_space_position+6]
        cards.append(card)
        blank_space_position += 6

bingo_order = [int(number) for number in bingo_order.split(',')]
```

```python
from dataclasses import dataclass

@dataclass
class BingoNumber:
    number: int = 0
    status: bool = False

class BingoCard:
    def __init__(self, input_card):
        self.card = [[BingoNumber(item) for item in row] 
                     for row in formatted_card]
        self.bingo_status = False
        
    def mark_number(self, number):
        for i, row in enumerate(self.card):
            for j, item in enumerate(row):
                if item.number == number:
                    self.card[i][j].status = True
                    
    def get_sum_of_unmarked(self):
        unmarked = [[item.number for item in row if not item.status] for row in self.card]
        sum_of_unmarked = sum([item for row in unmarked for item in row])
        return sum_of_unmarked
    
    def check_for_bingo(self):
        check_counter = 0
        while not self.bingo_status:
            for row in self.card:
                check_counter += 1
                status = [item.status for item in row]
                if all(status):
                    self.bingo_status = True
                    break
            for i in range(5):
                check_counter += 1
                status = [self.card[j][i].status for j in range(5)]
                if all(status):
                    self.bingo_status = True
                    break
            if check_counter == 10:
                break
                
        return self.bingo_status
```

```python
bingo_cards = []

for card in cards:
    formatted_card = [[int(item) for item in row.split()] for row in card]
    card = BingoCard(formatted_card)
    bingo_cards.append(card)

test_card = bingo_cards[1]
for mark in bingo_order:
    test_card.mark_number(mark)
    test_for_bingo = test_card.check_for_bingo()
    print(mark, test_for_bingo)
```

```python
bingo_cards = []

for card in cards:
    formatted_card = [[int(item) for item in row.split()] for row in card]
    card = BingoCard(formatted_card)
    bingo_cards.append(card)
    
bingo_was_reached = False
for mark in bingo_order:
    for card_number, card in enumerate(bingo_cards):
        card.mark_number(mark)
        test_for_bingo = card.check_for_bingo()
        if test_for_bingo:
            bingo_was_reached = True
            winner_card = card_number
            break
    if bingo_was_reached:
        print(f"Reached bingo with card number {winner_card}, number {mark}")
        print(bingo_cards[winner_card].get_sum_of_unmarked() * mark)
        break
        
```
